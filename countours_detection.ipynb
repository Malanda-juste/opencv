```python
import cv2
import numpy as np
```

```python
img = cv2.imread("./data/camera.jpeg",1)
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

thresh = cv2.adaptiveThreshold(gray_img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 253, 1)
cv2.imshow("thresh", thresh)

contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

img_2 = img.copy()
cv2.drawContours(img_2, contours, -1, (255,0,0), 2)
cv2.imshow("img_2",img_2)


cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
plt.hist(gray_img)
```

```python
h, w, c = img.shape
img3 = np.zeros([h,w,1], "uint8")

cv2.drawContours(img3, contours, -1, (255,0,0), 2)
cv2.imshow("img_2",img3)


cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Appliquer un flou gaussien pour réduire le bruit
blurred_image = cv2.GaussianBlur(img, (5, 5), 0)

# Appliquer la détection de contours avec Canny
edges = cv2.Canny(blurred_image, 50, 150)

# Afficher le résultat
plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.title('Original')
plt.imshow(img, cmap='gray')
plt.subplot(1, 2, 2)
plt.title('Edges')
plt.imshow(edges, cmap='gray')
plt.show()
```
