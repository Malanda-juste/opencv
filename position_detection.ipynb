```python
import mediapipe as mp
import cv2
```

```python
draw_tools = mp.solutions.drawing_utils
posEstimation = mp.solutions.pose
pose = posEstimation.Pose()
```

```python
cap = cv2.VideoCapture("./data/IL JONGLE AVEC DEUX BALLONS.mp4")
```

```python
while(True):
    ret, frame = cap.read() #lire la video
    
    #frame = cv2.resize(frame, (0,0), fx=0.9,fy=0.9) #redimensionne la video
    framRB = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    rs = pose.process(framRB)
    
    if rs.pose_landmarks:
        draw_tools.draw_landmarks(frame, rs.pose_landmarks, posEstimation.POSE_CONNECTIONS)
    print(rs.pose_landmarks)
    cv2.imshow("Image",frame) 
    cv2.moveWindow("Image", 100,10) 
    
    ch = cv2.waitKey(1)
    if ch & 0xFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```
